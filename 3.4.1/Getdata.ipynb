```python
import pandas as pd

data = pd.read_csv('afx.csv')
```

```python
test_list = data.to_dict('records')
```

```python
max_dict = {k: max(d[k] for d in test_list) for k in test_list[0].keys()}
 
print("Highest Value of Stock - ", max_dict['High'])
```

```python
min_dict = {k: min(d[k] for d in test_list) for k in test_list[0].keys()}
 
print("Lowest Value of Stock - ", min_dict['Low'])
```

```python
print("Highest Value of Closig Price - ", max_dict['Close'],max_dict['Date'])
```

```python
print("Lowest Value of Closig Price - ", min_dict['Close'],min_dict['Date'])
```
